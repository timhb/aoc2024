```python
import re
from time import sleep

def find_pos(robot, max_x, max_y, cycles):
    x, y, dx, dy = robot
    return ((x + cycles * dx) % max_x + max_x) % max_x, ((y + cycles * dy) % max_y + max_y) % max_y

def print_map(robots, max_x, max_y):
    map = [["." for _ in range(max_x)] for _ in range(max_y)]
    for robot in robots:
        map[robot[1]][robot[0]] = "#"
    for line in map:
        print("".join(line))

def main():

    max_x, max_y = 101, 103
    with open("robots.txt", "r") as f:
        file = f.read()
        pattern = r"p=(\d+),(\d+) v=([-\d]+),([-\d]+)"
        matches = re.findall(pattern, file)
        robots = []
        for match in matches:
            pos_x, pos_y, dx, dy = map(int, match)
            robots.append([pos_x, pos_y, dx, dy])

    for i in range(10000):
        robots_pos = []
        for robot in robots:
            robots_pos.append(find_pos(robot, max_x, max_y, i))
        # There will probably be no duplicates (lucky)
        if len(set(robots_pos)) == len(robots_pos):
            print(i)
            print_map(robots_pos, max_x, max_y)

    # amount_per_quadrant = [0] * 4
    # for robot_pos in robots_pos:
    #     x, y = robot_pos
    #     half_x, half_y = max_x // 2, max_y // 2
    #     if x != half_x and y != half_y:
    #         amount_per_quadrant[2*(x < half_x) + (y < half_y)] += 1
    #
    # total = 1
    #
    # for i in range(4):
    #     total *= amount_per_quadrant[i]
    # print(total)

if __name__ == "__main__":
    main()
```
