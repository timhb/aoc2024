```python
from copy import deepcopy

def in_limits(pos, map):
    if 0 <= pos[0] < len(map[0]) and 0 <= pos[1] < len(map):
        return True
    return False

def flood_fill(pos, map, group):
    x,y = pos[0], pos[1]
    value = map[y][x]
    group.append((x,y))
    neighbors = {(x, y + 1), (x, y - 1), (x + 1, y), (x - 1, y)}
    for neighbor in neighbors:
        if in_limits(neighbor, map):
            neighbor_value = map[neighbor[1]][neighbor[0]]
            if neighbor_value == value and neighbor not in group:
                flood_fill(neighbor, map, group)

def calc_outer_connections(group, map):
    outer_connections = 0
    for field in group:
        x, y = field[0], field[1]
        dx, dy = -1, 0
        for _ in range(4):
            if in_limits((dx+x,dy+y), map):
                if map[dy+y][dx+x] != map[y][x]:
                    outer_connections += 1
            else:
                outer_connections += 1
            dx, dy = -dy, dx
    return outer_connections

def calc_lines(group, map):
    # Idea: instead of counting lines count corner
    # - When do we have a corner?
    corners = 0
    for field in group:
        x,y = field[0], field[1]
        dx, dy = -1, 1
        for _ in range(4):
            empty_fields = []
            for dir in [(dx, 0), (dx, dy), (0, dy)]:
                if in_limits((x + dir[0], y + dir[1]), map):
                    if (x + dir[0], y + dir[1]) not in group:
                        empty_fields.append(1)
                    else:
                        empty_fields.append(0)
                else:
                    empty_fields.append(1)
            if empty_fields == [0, 1, 0]:
                corners += 1
                print(field)
            elif empty_fields == [1, 1, 1]:
                corners += 1
                print(field)
            elif empty_fields == [1, 0, 1]:
                corners += 1
                print(field)
            dx, dy = -dy, dx
    return corners

def print_groups(groups, map):
    map_cpy = deepcopy(map)
    group_idx = 0
    for group in groups:
        for field in group:
            map_cpy[field[1]][field[0]] = group_idx
        group_idx += 1

    for line in map_cpy:
        print(" ".join(map(str, line)))


def main():
    # Load map
    with open("garden_plots.txt", "r") as file:
        map = [list(line.strip("\n")) for line in file]

    # Group cells to groups (list(list))
    visited = set()
    groups = []
    for y, line in enumerate(map):
        for x, cell in enumerate(line):
            if (x,y) not in visited:
                group = []
                flood_fill((x,y), map, group)
                for element in group:
                    visited.add(element)
                groups.append(group)

    # create list for each group with total_connections to outer land plus amount of fields
    prices = []
    for group in groups:
        total_amount = len(group)
        # total_outer_connections = calc_outer_connections(group, map)
        total_outer_connections = calc_lines(group, map)
        prices.append((total_amount, total_outer_connections))

    # list total price
    total_price = 0
    for price in prices:
        total_price += price[0] * price[1]
    print(total_price)

if __name__ == "__main__":
    main()
```
