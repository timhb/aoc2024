### Load data

```python
from decorator import append
from jsonschema.validators import extend

data = open("safety_manual.txt", "r")

rules = []
line = ""
while True:
    line = data.readline()
    if line == "\n":
        break
    rules_split = line.split("|")
    rules.append((rules_split[0], rules_split[1].removesuffix("\n")))

rules_list = [[] for _ in range(100)]
for rule in rules:
    rules_list[int(rule[0])].append(rule[1])


orderings = []
while True:
    line = data.readline()
    if not line:
        break
    ordering_split = line.split(",")
    ordering_split[len(ordering_split)-1] = ordering_split[len(ordering_split)-1].removesuffix("\n")
    orderings.append(ordering_split)

orderings
```

### Check data

```python
sum = 0

def check_for_occurence(list1, list2):
    for idx, element in enumerate(list1):
        if list2.count(element):
            return True
    return False

false_orderings = []
for ordering in orderings:
    is_correct = True
    for idx, page in enumerate(ordering):
        if check_for_occurence(ordering[:idx], rules_list[int(page)]):
            is_correct = False
    if is_correct:
        sum += int(ordering[int(len(ordering)/2)])
    else:
        false_orderings.append(ordering)

false_orderings
```

### Part Two: Repair false orderings

```python
new_orderings = []
for ordering in false_orderings:
    for start, page in enumerate(ordering):
        for current in range(start, 0, -1):
            if check_for_occurence(ordering[:current], rules_list[int(page)]):
                ordering[current], ordering[current-1] = ordering[current-1], ordering[current]
            else:
                break

sum = 0

for ordering in false_orderings:
    sum += int(ordering[int(len(ordering)/2)])

sum
```
