```python
import math
data = open("corrupted_data.txt", "r")

first_par = 0
second_par = 0

result = 0
enable = True

for line in data:
    for i in range(len(line)):
        if line[i:i+4] == "do()":
            enable = True
        elif line[i:i+7] == "don't()":
            enable = False
        if enable and line[i:i+4] == "mul(":
            max_line = line[i+4:min(len(line), i+12)]
            first = ""
            second = ""

            k = 0
            while k < 3 and max_line[k].isdigit():
                first += max_line[k]
                k+= 1
            if max_line[k] != ",":
                continue
            k += 1
            j = 0
            while j < 3 and max_line[k + j].isdigit():
                second += max_line[k + j]
                j += 1
            if max_line[k+j] != ")":
                continue
            if 0 < len(first) < 4 and 0 < len(second) < 4:
                if 0 < int(second) < 1000 and 0 < int(first) < 1000:
                    result += int(second) * int(first)
                    print(first + " * " + second)
result
```
